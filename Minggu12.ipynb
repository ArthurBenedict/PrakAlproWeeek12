```python
n = int(input('Masukkan jumlah kategori: '))

data_aplikasi = {}

for i in range(n):
    nama_kategori = input(f"Masukkan nama kategori ke-{i+1}: ")
    aplikasi = []
    print(f"Masukkan 5 nama aplikasi untuk kategori '{nama_kategori}':")
    for j in range(5):
        nama_aplikasi = input(f"Aplikasi ke-{j+1}: ")
        aplikasi.append(nama_aplikasi)
    data_aplikasi[nama_kategori] = aplikasi
    
print("\nDaftar nama aplikasi di setiap kategori:")
for kategori, aplikasi in data_aplikasi.items():
    print(f"{kategori}: {', '.join(aplikasi)}")

daftar_aplikasi_list = []

for aplikasi in data_aplikasi.values():
    daftar_aplikasi_list.append(set(aplikasi))


hasil = daftar_aplikasi_list[0]
for i in range(1, len(daftar_aplikasi_list)):
    hasil = hasil&daftar_aplikasi_list[i]
print(f"aplikasi yang muncul disemua kategori: {hasil}")

hasill = daftar_aplikasi_list[0]
for i in range(1, len(daftar_aplikasi_list)):
    hasill = hasill^daftar_aplikasi_list[i]

iyah = daftar_aplikasi_list[0]
for i in range(1,len(daftar_aplikasi_list)):
    iyah.update(iyah&daftar_aplikasi_list[i])
for i in iyah:
    if i in hasill:
        hasill.remove(i)
print(hasill)

print("Aplikasi yang hanya muncul di satu kategori saja:", hasill)

apl = list()
dua_kategori=set()

if n > 2:
    for i in data_aplikasi.values():
        apl+=i
    for i in apl:
        if apl.count(i)==2:
            dua_kategori.add(i)
    print("Aplikasi yang muncul tepat di dua kategori:", dua_kategori)
```

```python
def demo():
    Lst = [1, 2, 2, 3, 4, 4, 5]
    print(f'list sebelum di konversi : {Lst}')
    St = set(Lst)
    print(f'list setelah di konversi menjadi set: {St}')
    print(f'set sebelum di konversi : {St}')
    Lst2 = set(St)
    print(f'set setelah di konversi menjadi list : {Lst2}')

    Tple = [1, 2, 2, 3, 4, 4, 5]
    print(f"tuple sebelum di konversi : {Tple}")
    St2 = set(Tple)
    print(f"tuple setelah di konversi menjadi set: {St2}")
    print(f'set sebelum di konversi : {St2}')
    Tple2 = tuple(St2)
    print(f"tuple setelah di konversi menjadi tuple : {Tple2}")

demo()
```

```python
def file_ke_set(filename):
    try:
        with open(filename, 'r') as file:
            isifile = file.read().lower() 
            kata = set(isifile.split())   
        return kata
    except FileNotFoundError:
        print(f"Error! File '{filename}' tidak ditemukan.")
        return None
    except IOError:
        print(f"Error! File '{filename}' tidak bisa dibaca.")
        return None

def main():
    file1 = input("Masukkan nama file pertama: ")
    file2 = input("Masukkan nama file kedua: ")
    katafile1 = file_ke_set(file1)
    katafile2 = file_ke_set(file2)
    
    if katafile1 is not None and katafile2 is not None:
        seringmuncul = katafile1.intersection(katafile2)
        
        if seringmuncul:
            print("\nKata yang muncul di kedua file tersebut adalah:")
            for kata in seringmuncul:
                print(kata)
        else:
            print("\nTidak ada kata yang sama pada kedua file.")
    else:
        print("Proses dihentikan karena ada file yang tidak terbaca.")

if __name__ == "__main__":
    main()
```
